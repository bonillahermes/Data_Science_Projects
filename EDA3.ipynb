# Calcular las medidas estadisticas descriptivas

```python
df.describe()
```

count: Es el número de observaciones no nulas. Esto puede ser útil para identificar si hay valores nulos en los datos.

mean: Es la media (promedio) de los datos. Es útil para obtener una "idea general" de los valores de un conjunto de datos, pero puede ser influenciada por valores extremadamente altos o bajos (outliers).

std: Es la desviación estándar, que mide la cantidad de variación o dispersión de un conjunto de valores. Una baja desviación estándar indica que los valores tienden a estar cerca de la media, mientras que una alta desviación estándar indica que los valores están dispersos en un rango más amplio.

min: Es el valor mínimo en el conjunto de datos.

25%: Es el primer cuartil, o el valor por debajo del cual se encuentra el 25% de los datos.

50%: Es la mediana, o el valor que separa el conjunto de datos en dos mitades iguales. El 50% de los datos son menores que la mediana, y el 50% son mayores.

75%: Es el tercer cuartil, o el valor por debajo del cual se encuentra el 75% de los datos.

max: Es el valor máximo en el conjunto de datos.

# **HISTOGRAMAS**

Distribución general para algunas variables de interés.

```python
# Crear una lista de las variables de interes
variables_estudio = ['Adolescent fertility rate (births per 1,000 women ages 15-19)',
                         'CO2 emissions (metric tons per capita)',
                         'GDP (current US$)']

# Crear subplots para cada variable
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Dibujar la distribucion de cada variable
for i, variable in enumerate(variables_estudio):
    sns.histplot(df_2019[variable], kde=True, ax=axes[i])
    axes[i].set_title(f'Distribución de {variable}', fontsize=10)
    axes[i].set_xlabel(variable, fontsize=8)
    axes[i].set_ylabel('Frecuencia', fontsize=8)

plt.tight_layout()
plt.show()
```

# **Interpretación de los diagramas de cajas**

Mediana (Q2): La línea horizontal dentro del recuadro representa la mediana de los datos. Es el valor que separa la mitad superior de la mitad inferior del conjunto de datos. Nos da un valor central de los datos.

Rango intercuartil (IQR): La altura de la caja representa el rango intercuartílico (IQR). El IQR es el rango entre el primer cuartil (Q1, percentil 25) y el tercer cuartil (Q3, percentil 75). Representa el 50% medio de los datos y nos da una idea de la dispersión de los datos.

Bigotes: Los "bigotes" o las líneas que se extienden desde la caja muestran el rango de los datos. Cualquier punto de datos fuera de este rango se considera normalmente un valor atípico.

Valores atípicos: Los puntos individuales trazados más allá de los bigotes son valores atípicos. Son valores que caen fuera del rango típico (más de 1,5 veces el IQR de los cuartiles) y pueden indicar variabilidad en los datos o errores potenciales.

Para la variable **Expectativa de vida**, esta podría ser la interpretación:
    
Media: 72.19 años. Esto indica que, en promedio, la expectativa de vida al nacer en todo el mundo en 2019 fue de aproximadamente 72.19 años.

Desviación estándar: 7.78 años. Esto indica que la mayoría de los países tienen una expectativa de vida al nacer que se desvía en 7.78 años de la media.

Mínimo: 49 años. Esto indica que el país con la menor expectativa de vida al nacer en 2019 tenía una expectativa de vida de 49 años.

25%: 66.33 años. Esto indica que el 25% de los países en 2019 tenía una expectativa de vida al nacer menor a 66.33 años.

Mediana (50%): 73.40 años. Esto indica que la mitad de los países en 2019 tenía una expectativa de vida al nacer menor a 73.4 años y la otra mitad tenía una expectativa de vida mayor a 73.4 años.

75%: 78.39 años. Esto indica que el 75% de los países en 2019 tenía una expectativa de vida al nacer inferior a 78.39 años.

Máximo: 85.50 años. Esto indica que el país con la mayor expectativa de vida al nacer en 2019 tenía una expectativa de vida de 85.5 años.
